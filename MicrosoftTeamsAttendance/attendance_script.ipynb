```python
import pandas as pd
from datetime import datetime
```

```python
attendance_df = pd.read_csv("raw_attendance.csv")
```

```python
attendance_df = attendance_df.sort_values(by = ["Full Name", "Timestamp"])
attendance_df = attendance_df.reset_index(drop = True)
attendance_df
```

```python
length = len(attendance_df.index)
format1 = " %H:%M:%S %p"
```

```python
output_df = pd.DataFrame()
```

```python
for i in range(0, length):
    
        full_name = attendance_df.loc[i, "Full Name"]
        
        if len(full_name.split("_")) == 2:
            name, entry_no = full_name.split("_")
        else:
            name = full_name
            entry_no = ""
        
        date, time = attendance_df.loc[i, "Timestamp"].split(",")
        action = attendance_df.loc[i, "User Action"]
        
        if(i == 0 ):
            timestamp_join = []
            timestamp_left = []
            timestamp_duration = []
            timestamp_join.append(time)
            extracted_name = name
            extracted_entry = entry_no
            
        elif(attendance_df.loc[i-1, "Full Name"] == attendance_df.loc[i, "Full Name"]):
            if(action == "Joined" or action == "Joined Before"):
                timestamp_join.append(time)
            elif(action == "Left"):
                timestamp_left.append(time)
            
        else:
            
            if(len(timestamp_join) != len(timestamp_left)):
                timestamp_left.append(" 12:30:00 PM")
            
            for x in range(0, len(timestamp_join)):
                t1 = datetime.strptime(timestamp_join[x], format1)
                t2 = datetime.strptime(timestamp_left[x], format1)
                difference = t2 - t1
                timestamp_duration.append(difference)
            
            timestamp = ""
            for x in range(0, len(timestamp_join)):
                timestamp = timestamp + '(' + str(timestamp_join[x]) + ', ' + str(timestamp_left[x]) + ', ' + str(timestamp_duration[x]) + ')'
                if(x != 0):
                    timestamp_duration[0] += timestamp_duration[x] 
            
            if(timestamp_duration[0].seconds >= 2400):
                status = "Present"
            else:
                status = "Absent"
            
            new_row = pd.Series([extracted_name, extracted_entry, timestamp, timestamp_duration[0], status])
            
            new_row_df = pd.DataFrame([new_row])
            
            output_df = pd.concat([output_df, new_row_df], ignore_index = True)
            
            timestamp_join = []
            timestamp_left = []
            timestamp_duration = []
            timestamp_join.append(time)
            extracted_name = name
            extracted_entry = entry_no
```

```python
if(len(timestamp_join) != len(timestamp_left)):
    timestamp_left.append(" 12:30:00 PM")

for x in range(0, len(timestamp_join)):
    t1 = datetime.strptime(timestamp_join[x], format1)
    t2 = datetime.strptime(timestamp_left[x], format1)
    difference = t2 - t1
    timestamp_duration.append(difference)

timestamp = ""
for x in range(0, len(timestamp_join)):
    timestamp = timestamp + '(' + str(timestamp_join[x]) + ', ' + str(timestamp_left[x]) + ', ' + str(timestamp_duration[x]) + ')'
    if(x != 0):
        timestamp_duration[0] += timestamp_duration[x] 

if(timestamp_duration[0].seconds >= 2400):
    status = "Present"
else:
    status = "Absent"

new_row = pd.Series([extracted_name, extracted_entry, timestamp, timestamp_duration[0], status])

new_row_df = pd.DataFrame([new_row])

output_df = pd.concat([output_df, new_row_df], ignore_index = True)
```

```python
output_df = output_df.rename(columns = {0: "Name", 1: "Entry No.", 2: "TimeStamp", 3: "Total Time", 4: "Status"})
output_df
```

```python
output_df.to_csv("processed_attendance.csv")
```
